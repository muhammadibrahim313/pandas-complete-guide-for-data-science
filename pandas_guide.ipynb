# 1. Create a DataFrame

4. Viewing Data
Pandas provides functions to view and inspect data stored in Series and DataFrame objects.

4.1 Head and Tail

```python
# Create a DataFrame
df = pd.DataFrame({'A': range(10), 'B': range(10, 20)})

# View the first few rows
print(df.head())

print('---------------')

# View the last few rows
print(df.tail())
```

5. Selection
Selection in pandas allows you to access subsets of data from Series and DataFrame objects.

5.1 Selection by Label

```python
# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']}, index=['X', 'Y', 'Z'])

# Select a single column by label
print(df.loc['X'])

# Select a single value by label
print(df.loc['X', 'A'])

# Select a subset of rows and columns by label
print(df.loc[['X', 'Y'], 'A':'B'])
```

5.2 Selection by Position


```python

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']})
print(df)


print('--------------------------')

# Select a single value by position
print(df.iloc[0, 1])

print('--------------------------')

# Select a subset of rows and columns by position
print(df.iloc[0:2, :])
```

6. Missing Data
Dealing with missing data is a crucial aspect of data analysis. Pandas provides methods to handle missing data effectively.

6.1 Identifying Missing Values

```python


# Create a DataFrame with missing values
df = pd.DataFrame({'A': [1, np.nan, 3], 'B': [np.nan, 5, 6]})

# Check for missing values
print(df.isnull())
```

6.2 Handling Missing Values


```python
import pandas as pd
import numpy as np

# Create a DataFrame with missing values
df = pd.DataFrame({'A': [1, np.nan, 3], 'B': [np.nan, 5, 6]})

# Drop rows with missing values
print(df.dropna())

print('-----------')

# Fill missing values with a specific value
print(df.fillna(0))
```

7. Operations
Pandas provides powerful operations for data manipulation, including arithmetic operations, aggregation, and more.

7.1 Arithmetic Operations

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})

# Add two DataFrames
print(df1 + df2)

print('--------------')

# Multiply two DataFrames
print(df1 * df2)
```

7.2 Aggregation

```python
# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]})

# Calculate mean of each column
print(df.mean())

print('------------')

# Calculate sum of each row
print(df.sum(axis=1))
```

8. Merge
Merging in pandas allows you to combine data from different DataFrames based on common columns or indices.

8.1 Merge by Column

```python
# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'C': [10, 11, 12]})

# Merge based on a common column
merged_df = pd.merge(df1, df2, on='A')
print(merged_df)
merged_df.head()
```

8.2 Merge by Index

```python
# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3]}, index=['X', 'Y', 'Z'])
df2 = pd.DataFrame({'B': [4, 5, 6]}, index=['X', 'Y', 'Z'])

# Merge based on index
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
print(merged_df)
```




1. Grouping
Grouping in pandas involves splitting the data into groups based on some criteria, applying a function to each group, and combining the results into a new data structure.

9.1 Splitting the Data into Groups

```python
# Create a DataFrame
data = {'Team': ['A', 'B', 'A', 'B', 'A'],
        'Player': ['P1', 'P2', 'P3', 'P4', 'P5'],
        'Points': [10, 20, 15, 25, 30]}
df = pd.DataFrame(data)

# Group by 'Team'
grouped = df.groupby('Team')

# Display groups
for name, group in grouped:
    print(name)
    print(group)
    print()
```

9.2 Applying Functions to Groups

```python
# Calculate the sum of points for each team
print(grouped.sum())
```

10. Reshaping
Reshaping in pandas involves transforming the layout of your data, such as pivoting, stacking, and melting, to better suit your analysis needs.

10.1 Pivoting

```python
# Pivot the DataFrame
pivot_df = df.pivot(index='Team', columns='Player', values='Points')
print(pivot_df)
```

10.2 Stacking and Unstacking

```python
# Stack the DataFrame
stacked_df = df.set_index(['Team', 'Player']).stack()
print(stacked_df)
```



11.  Time Series
Time series analysis is a crucial aspect of data analysis, especially when dealing with data indexed by time or date.

11.1 Working with Timestamps

```python
# Create a Series with timestamps as index
dates = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03'])
values = [100, 200, 300]
ts = pd.Series(values, index=dates)

print(ts)
```

11.2 Resampling Time Series Data

```python
# Resample the time series to monthly frequency
monthly_ts = ts.resample('M').mean()
print(monthly_ts)
```

12. Categorical Data
Categorical data represents data that can take on a limited, fixed number of possible values.

12.1 Creating Categorical Data


```python

# Create a Series with categorical data
s = pd.Series(['cat', 'dog', 'dog', 'cat', 'dog'], dtype='category')
print(s)
```

12.2 Working with Categorical Data

```python
# Count occurrences of each category
print(s.value_counts())

# Rename categories
s = s.cat.rename_categories({'cat': 'kitty', 'dog': 'puppy'})
print(s)
```


13.  Plotting
Pandas provides convenient methods for plotting data directly from DataFrames and Series objects.

13.1 Basic Plotting with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'Year': [2018, 2019, 2020, 2021, 2022],
        'Sales': [100, 150, 200, 250, 300]}
df = pd.DataFrame(data)

# Plot the data
df.plot(x='Year', y='Sales', kind='line')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

13.2 Other Plotting Options


```python
# Plot a bar chart

df.plot(x='Year', y='Sales', kind='bar')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(axis='y')
plt.show()
```

21. Group By: Split-Apply-Combine
Grouping data with pandas allows you to perform split-apply-combine operations efficiently.

21.1 Splitting an Object into Groups

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'Category': ['A', 'B', 'A', 'B'],
                   'Value': [10, 20, 30, 40]})

# Group by 'Category'
grouped = df.groupby('Category')

# Display groups
for name, group in grouped:
    print(name)
    print(group)
    print()
```

21.2 Aggregation

```python
# Calculate the sum for each group
print(grouped.sum())
```

22. Windowing Operations
Window functions in pandas allow you to perform calculations over a sliding window of data.

22.1 Rolling Window

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'Value': [10, 20, 30, 40, 50]})

# Calculate the rolling mean
rolling_mean = df['Value'].rolling(window=2).mean()
print(rolling_mean)
```

22.2 Expanding Window

```python
# Calculate the expanding sum
expanding_sum = df['Value'].expanding().sum()
print(expanding_sum)
```
